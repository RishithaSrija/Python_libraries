```python
path='/content/OnlineRetail.csv'
df=pd.read_csv('OnlineRetail.csv')


```

```python
#1
import pandas as pd
print("Top 5 rows:")
display(df.head())
print("\nLast 5 rows:")
display(df.tail())
print("\nDataFrame Info:")
df.info()
print("\nShape of DataFrame:", df.shape)
print("Data types: ",df.dtypes)
print()
#2 missing data handling

print("Missing values: ",df.isnull().sum())
print()

df=df.dropna(subset=['CustomerID'])

df['Description']=df['Description'].fillna('Unknown')
df.fillna('Unknown', inplace=True)
print("Missing values after cleaning:\n", df.isnull().sum())
print("\nBottom rows after cleaning:")
print(df.tail().to_string())

#3
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'],errors='coerce')
#drop where parsing failed
df=df.dropna(subset=['InvoiceDate'])
print("\nBottom rows after parsing:")
print(df.tail().to_string())
print()

#4
negative_qty = df[df['Quantity'] < 0]
print("Transactions with negative Quantity (returns):\n", negative_qty.head())
print("Total returns:", negative_qty.shape[0])
print()
df['IsReturn'] = df['Quantity'] < 0
df = df[df['Quantity'] >= 0]

invalid_price=df[df['UnitPrice']<0]
print("Transactions with zero or negative UnitPrice:\n", invalid_price.head())
#5
duplicates = df[df.duplicated(subset=['InvoiceNo', 'StockCode'], keep=False)]
print("Duplicated invoice-item entries:\n", duplicates.head())
print("Total duplicates:", duplicates.shape[0])

df_cleaned = df.drop_duplicates(subset=['InvoiceNo', 'StockCode'], keep='first')

print("Shape before dropping duplicates:", df.shape)
print("Shape after dropping duplicates:", df_cleaned.shape)
```

```python
import pandas as pd
path = '/content/OnlineRetail.csv'
df = pd.read_csv(path,encoding='ISO-8859-1')
#6.filtering& sorting
sus_outliers=df[df['Quantity']>1000]
print("Filtering Suspicious transactions: ",sus_outliers.head())

df['Revenue']=df['UnitPrice']*df['Quantity']
top_revenue = df.nlargest(5,'Revenue')
print("\nTop 5 highest revenue orders:\n", top_revenue[['InvoiceNo', 'StockCode', 'Quantity', 'UnitPrice', 'Revenue']])
#7.Merge
product_info = pd.DataFrame({
    'StockCode': ['85123A', '71053', '84406B'],
    'Category': ['Home', 'Office', 'Garden'],
    'Supplier': ['SupplierA', 'SupplierB', 'SupplierC']
})
df_merged = pd.merge(df, product_info, on='StockCode', how='left')

print("\nData after merging with product_info:\n", df_merged.head())
#8
df['Revenue']=df['UnitPrice']*df['Quantity']
ct_summary=df.groupby('Country').agg(
    Total_Revenue=('Revenue','sum'),
    Mean_Revenue=('Revenue','mean'),
    Order_count=('InvoiceNo','count'),
).sort_values(by='Total_Revenue',ascending=False)
print("\n Revenue Country Summary:\n", ct_summary.head())

bins=[0,5,20,df['UnitPrice'].max()]
labels=['Low','Medium','High']
df['PriceCategory']=pd.cut(df['UnitPrice'],bins=bins,labels=labels)

price_summary=df.groupby('PriceCategory').agg(
    Total_purchases=('InvoiceNo','count'),
    Avg_Revenue=('Revenue','mean')
)
print("\nPrice Category Summary:\n",price_summary)

#9
numeric_cols=['Quantity','UnitPrice','Revenue']
corr_matrix=df[numeric_cols].corr()
print("\nCorrelation Matrix:\n",corr_matrix)
print()

corr_pairs=corr_matrix.unstack().reset_index()
corr_pairs.columns=['Feature1','Feature2','Correlation']

corr_pairs=corr_pairs[corr_pairs['Feature1']!=corr_pairs['Feature2']]
strong_corr=corr_pairs[corr_pairs['Correlation']>0.7]
print("\nStrong Positive Correlation Pairs:\n",strong_corr)

#10
df.rename(columns={'StackCode':'Product_ID','UnitPrice':'Price'},inplace=True)
df['Desc_lower']=df['Description'].str.lower()
glass_prod=df[df['Desc_lower'].str.contains('glass',na=False)]
print("\nProducts with 'Glass' in Description:\n",glass_prod.head())


```
